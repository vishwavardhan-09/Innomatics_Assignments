```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print(" Recharge amount must be at least ₹50")
        return False
    elif amount not in valid_plans:
        print(" Invalid plan selected")
        return False
    else:
        print(" Recharge Successful")
        return True


# Retry option using while loop
while True:
    amount = int(input("Enter recharge amount: "))

    if validate_recharge(amount):
        break
    else:
        print("Please try again...\n")
```

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}:  Reorder Alert")
        else:
            print(f"{product}:  Stock OK")


# Example dictionary
inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 8,
    "Flour": 25
}

check_inventory(inventory)
```

```python
def process_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Average: {average} - Pass"
    else:
        return f"Average: {average} - Fail"


# Example
student_marks = [60, 45, 70, 55]
print(process_result(student_marks))
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)

    if peak.lower() == "yes":
        fare += fare * 0.25

    return fare


# Retry option
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    fare = calculate_fare(distance, peak)
    print(f"Estimated Fare: ₹{fare}")

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
def check_attendance(attendance_list):
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    if percentage >= 75:
        return f"Attendance: {percentage}% - Eligible"
    else:
        return f"Attendance: {percentage}% - Not Eligible"


# Example
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
print(check_attendance(attendance))
```

```python
def check_password(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "❌ Password too short"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return " Strong Password"
    else:
        return " Weak Password"


# Example
pwd = input("Enter password: ")
print(check_password(pwd))
```

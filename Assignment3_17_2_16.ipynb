```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

# Find all employees with highest score (handle ties)
top_performers = [name for name, score in employees.items() if score == highest_score]

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation (if any)
import string
query = query.translate(str.maketrans('', '', string.punctuation))

# Split into words
words = query.split()

# Count frequency
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# Display words searched more than once
result = {word: count for word, count in word_count.items() if count > 1}

print("Repeated Keywords:", result)
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:   # Check if even
        valid_readings.append((hour, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Extract domains
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculate percentages
total_emails = len(emails)

print("Email Domain Usage:")
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
average_sales = sum(sales) / len(sales)

print("Sales Spikes Detected:")

for day, value in enumerate(sales, start=1):
    if value > average_sales * 1.30:   # More than 30% above average
        print(f"Day {day}: {value}")
```

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

# Count occurrences
for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1

print("Duplicate User IDs:")

for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
